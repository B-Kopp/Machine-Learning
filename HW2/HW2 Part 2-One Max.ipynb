```python
import mlrose
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
best_state = []
best_fitness = []
space = np.linspace(5,100,20)

for N in space:
    fitness = mlrose.OneMax()
    problem = mlrose.DiscreteOpt(length = N, fitness_fn = fitness, maximize = True, max_val = 2)
    # Set random seed
    np.random.seed(1)
    # Solve problem using random hill climbing
    state, fitness = mlrose.random_hill_climb(problem, max_attempts = 100, max_iters = 5000)
    best_state.append(state)
    best_fitness.append(fitness)
    
plt.figure()
plt.suptitle("One Max Randomized Hill Climbing Fitness Curve")
plt.title("No restarts, 5,000 iterations")
plt.xlabel("N")
plt.ylabel("Fitness")

plt.grid()
plt.plot(space, best_fitness)
plt.show()
```

```python
start = time.time()

fitness = mlrose.OneMax()
problem = mlrose.DiscreteOpt(length = 100, fitness_fn = fitness, maximize = True, max_val = 2)
# Set random seed
np.random.seed(1)
# Solve problem using random hill climbing
state, fitness, curve = mlrose.random_hill_climb(problem, max_attempts = 100, max_iters = 5000, restarts=1, curve=True)

end = time.time()

print("Random Hill Climbing Algorithm Time:", round(end - start,4))
print("Random Hill Climbing Algorithm Fitness Score:", fitness)
    
plt.figure()
plt.suptitle("One Max Randomized Hill Climbing Fitness Curve")
plt.title("1 restart, N=100")
plt.xlabel("Iterations")
plt.ylabel("Fitness")

plt.grid()
plt.plot(np.linspace(1,len(curve),len(curve)), curve)
plt.show()
```

```python
best_state = []
best_fitness = []
space = np.linspace(5,100,20)

for N in space:
    # Define decay schedule
    schedule = mlrose.GeomDecay()
    fitness = mlrose.OneMax()
    problem = mlrose.DiscreteOpt(length = N, fitness_fn = fitness, maximize = True, max_val = 2)
    # Set random seed
    np.random.seed(1)
    # Solve problem using simulated annealing
    state, fitness = mlrose.simulated_annealing(problem, schedule=schedule, max_attempts = 100, max_iters = 5000)
    best_state.append(state)
    best_fitness.append(fitness)
    
plt.figure()
plt.suptitle("One Max Simulated Annealing Fitness Curve")
plt.title("Geometric Decay, 5,000 iterations")
plt.xlabel("N")
plt.ylabel("Fitness")

plt.grid()
plt.plot(space, best_fitness)
plt.show()
```

```python
best_state = []
best_fitness = []
space = [mlrose.ExpDecay(), mlrose.GeomDecay(), mlrose.ArithDecay()]

for N in space:
    # Define decay schedule
    schedule = N
    fitness = mlrose.OneMax()
    problem = mlrose.DiscreteOpt(length = 100, fitness_fn = fitness, maximize = True, max_val = 2)
    # Set random seed
    np.random.seed(1)
    # Solve problem using simulated annealing
    state, fitness = mlrose.simulated_annealing(problem, schedule=schedule, max_attempts = 100, max_iters = 5000)
    best_state.append(state)
    best_fitness.append(fitness)
    
names = ['ExpDecay', 'GeomDecay', 'ArithDecay']
y_pos = np.arange(len(names))
plt.figure()
plt.suptitle("One Max Simulated Annealing Fitness Curve")
plt.title("N=100, 5,000 iterations")
plt.xlabel("Decay Schedule")
plt.ylabel("Fitness")
plt.xticks(y_pos, names)

plt.ylim(80,105)
plt.bar(y_pos, best_fitness, align='center', alpha=0.5)
plt.show()
```
